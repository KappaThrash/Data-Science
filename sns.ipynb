```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_theme()
sns.set_style('whitegrid')

data = sns.load_dataset('tips')
data
```

```python
sns.scatterplot(x= 'tip', y= 'total_bill', data=data, hue='day',size='size',style='sex')
plt.tight_layout
plt.xlabel('Gorjeta')
plt.ylabel('Pagamento')
plt.title('RelaÃ§Ã£o entre Pagamento e Gorjeta')
plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
sns.countplot(x='sex', hue='smoker', data=data)
plt.tight_layout
plt.title('Quantidade de fumantes em relaÃ§Ã£o ao sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
```

```python
sns.countplot(x='time' ,data=data)
plt.title('Quantidade de clientes em cada dado horario')
plt.xlabel('Horario')
plt.ylabel('Quantidade')
```

```python
sns.kdeplot(x='total_bill', y='tip', data=data, fill=True)
```
